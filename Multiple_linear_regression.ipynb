## Multiple Linear Regression

```python
import pandas as pd
df = pd.read_csv('SalaryMulti.csv')
df.head()
```

```python
# Seprating our input and output data
x = df[['Total Experience', 'Team Lead Experience', 'Project Manager Experience', 'Certifications']]
y = df['Salary']
```

```python
x.head()
```

```python
y.head()
```

```python
# Spllitting dataset into training and testing data (80/20)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Creating and fitting our model 
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)
model
```

```python
# Plotting x_training data
import seaborn as sns
sns.scatterplot(x_train)
```

```python
# Plotting y_training data
import seaborn as sns
sns.scatterplot(y_train)
```

```python
# Plotting x_testing data
import seaborn as sns
sns.scatterplot(x_test)
```

```python
# Plotting y_testing data
import seaborn as sns
sns.scatterplot(y_test)
```

```python
# Checking training and testing scores of our model 
print(f"Training score = {model.score(x_train,y_train)*100}%")
print(f"Testing score = {model.score(x_test,y_test)*100}%")
```

```python
# Checking coefficients of Multiple Linear Regression Expression
model.coef_
```

```python
# Checking Intersept 'b' of Multiple Linear Regression Expression
model.intercept_
```

```python
# Making Predictions 
model.predict([[5,4,6,2]])
```

```python
# Checking accuracy of our model 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Separating data into 80% training and 20% testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Creating and fitting the model
model = LinearRegression().fit(x_train, y_train)

# Predicting values on X_test
predicted_values = model.predict(x_test)

# Evaluating the model
r2 = r2_score(y_test, predicted_values)
mse = mean_squared_error(y_test, predicted_values)

print("R-squared:", r2)
print("Mean Squared Error:", mse)
```
